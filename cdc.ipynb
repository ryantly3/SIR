```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set
plt.style.use('ggplot')
sns.set_context("poster")
```

```python
df = pd.read_csv('FluView_StackedColumnChart_Data.csv')
```

```python
df = df.drop(['YEAR', 'WEEK','TOTAL SPECIMENS'], axis=1)
```

```python
plt.rcParams["figure.figsize"]=20,10
df.plot(kind='bar', stacked=True)
plt.gca().set_title('Influenza Positive Tests Reported to CDC 2017-2018')
plt.gca().set_xlabel('Week')
plt.gca().set_ylabel('Number of Positive Specimens')
```
